```python
# Updated 8/2/19

# Cell 1

print('\nBegin Part 1 of 7 - Imports\n')

import datetime as dt
import numpy as np
import pandas as pd
import time

start_time = dt.datetime.now()
print('Start time: ' + str(dt.datetime.now()))

print('\nFinished Part 1 of 7 - Imports.  Ready to move to next cell.')
```

```python
# Cell 2

# Set initial variables
print('\nBegin Part 2 of 7 - Set variables\n')

# Set initial_path
initial_path = '/home/pace/glucose/data_files/adolescent-2160-0-Dexcom-Cozmo-Basal/'

input_train_file_name_prefix = 'adolescent#002'
input_test_file_name_prefix = 'adolescent#002'

# Set train_file_name and test_file_name
train_file_name = input_train_file_name_prefix + '-train.csv'
test_file_name = input_test_file_name_prefix + '-test.csv'

# Set processed_train_file_name and processed_test_file_name
processed_train_file_name = input_train_file_name_prefix + '-train-processed.csv'
processed_test_file_name = input_test_file_name_prefix + '-test-processed.csv'

print('\nFinished Part 2 of 7 - Set variables.  Ready to move to next cell.')
```

```python
# Cell 3

# Process the files.  This includes the following:
# 1.  Open the train and test files and read them into dataframes
# 2.  Convert the Time field in both df from an object to a datetime so it can be manipulated
# 3.  Drop unnecessary columns in both df
# 4.  Split the Time field into date and hour in both df.  These will be used for grouping when the 
#     LSTM or ARIMA are run.
# 5.  Save the train and test dataframes to csv files

print('\nBegin processing files\n\n**********\n\n')

# Open the csv files and read them into dataframes
print('\nBegin Part 3 of 7 - Open input files and create dataframes\n')

# Read the train file into train_df
train_df = pd.read_csv(str(initial_path + train_file_name), index_col=False)
print('Train file opened')
print('train_df shape = ' + str(train_df.shape))

# Read the test file into test_df
test_df = pd.read_csv(str(initial_path + test_file_name), index_col=False)
print('Test file opened')
print('test_df shape = ' + str(test_df.shape))

print('\nFinished Part 3 of 7 - Open input files and create dataframes\n\n**********\n\n')

# Convert the Time field in both df from an object to a datetime so it can 
# be manipulated
print('\nBegin Part 4 of 7 - Convert Times\n')

train_df['Time'] = pd.to_datetime(train_df['Time'])
test_df['Time'] = pd.to_datetime(test_df['Time'])

print('\nFinished Part 4 of 7 - Convert Times\n\n**********\n\n')
   
# Drop unnecessary columns in both df
print('\nBegin Part 5 of 7 - Drop columns\n')

train_df.drop(['CGM', 'CHO', 'insulin', 'LBGI', 'HBGI', 'Risk'], axis=1, inplace=True)
test_df.drop(['CGM', 'CHO', 'insulin', 'LBGI', 'HBGI', 'Risk'], axis=1, inplace=True)

print('\nFinished Part 5 of 7 - Drop columns\n\n**********\n\n')

# Split the Time field into date and hour in both df.  These will be used for grouping when the 
# LSTM or ARIMA are run.
print('\nBegin Part 6 of 7 - Split time for train and test then save processed files\n')

print('Splitting time for train - this may take a few minutes\n')
for i in range(0, len(train_df)):
    train_df['Date'] = pd.DatetimeIndex(train_df['Time']).date
    train_df['Hour'] = pd.DatetimeIndex(train_df['Time']).hour
print('Finished splitting time for train.  Thanks for your patience.\n')

print('Splitting test - this may take a few minutes\n')
for i in range(0, len(test_df)):
    test_df['Date'] = pd.DatetimeIndex(test_df['Time']).date
    test_df['Hour'] = pd.DatetimeIndex(test_df['Time']).hour    
print('Finished splitting time for test.   Thanks for your patience.\n')

print('\nFinished Part 6 of 7 - Split time for train and test then save processed files\n\n**********\n\n')

# Save train_df and test_df to csv files
print('\nBegin Part 7 of 7 - Saving processed data frames to csv files')
train_df.to_csv(str(initial_path + processed_train_file_name), index=False)
test_df.to_csv(str(initial_path + processed_test_file_name), index=False)
print('\nFinished Part 7 of 7 - Saving processed data frames to csv files\n\n**********\n\n')

print('\nFiles are completely processed.')
```

```python
# Cell 4

end_time = dt.datetime.now()
print('End time: ' + str(dt.datetime.now()))

print('\n\nScript completed in ' + str((end_time-start_time).total_seconds()) + ' seconds.')
```
